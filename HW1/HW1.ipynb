Having a faster sample rate allows for better reconstruction. If the sample rate is too low, there can be information you can't reproduce from the original audio. This is why CD audio sounds better than telephone audio. The CD quality audio sample rate is 44.1 kHz because it tries to take into account the Nyquist frequency. Aliasing can occur from undersampling. The Nyquist frequency says to be over twice the frequency of the highest frequency you want to represent. Human's hearing range is 20Hz - 20 kHz. 20 kHz * 2 = 40 kHz. 44.1 kHz is above that. Additionally, signals must be put through a low-pass filter or else aliasing occurs. We make a wider transition band to prevent aliasing from occuring. Therefore, sampling at 44.1kHz gives us leaveway for a 2.05kHz transition band and be able to accurately record the frequencies in the upper range of human hearing.

#### 3. (one point) Make a function to generate sinewaves and a function  to plot an audio signal. Label the figure's axes appropriately.

```python
def make_sinewave(f, t, sr):
    """    
    Parameters
    ----------
    f:  float
        Frequency of sine wave
    t:  float
        Duration in seconds
    sr: int
        Sample rate
    
    Returns
    -------
    np.ndarray 
        Array of floats containing the signal
    """
    return np.sin([2 * np.pi * x * f / sr for x in range(int(sr * t))])

def plot_audio(x, sr, figsize=(16,4)):
    """
    A simple audio plotting function
    
    Parameters
    ----------
    x: np.ndarray
        Audio signal to plot
    sr: int
        Sample rate
    figsize: tuple
        A duple representing the figure size (xdim,ydim)
    """
    length = float(x.shape[0]) / sr
    t = np.linspace(0,length,x.shape[0])
    plt.figure(figsize=figsize)
    plt.plot(t, x)
    plt.ylabel('Amplitude')
    plt.xlabel('Time (s)')
    plt.show()
```

#### 5. (one point) Implement a simple function to express the Root Mean Squared Amplitude of a signal encoded as an array, in dB, with respect to a reference value.

```python
def rms_db(signal, reference):
    '''
    Parameters
    ----------
    signal:  np.ndarray 
        Array of floats containing the signal
    reference:  float
        A reference amplitude to compare to 

    Returns
    -------
    float
        a value in dB, given the reference value
    '''
    s = 0
    for val in signal:
        s += val*val
    sig_rms = np.sqrt(s * 1.0 / signal.shape[0])
    return 20 * np.log10(sig_rms / reference)
```

#### 7. (one point) Implement a Fourier transform function and an inverse Fourier transform function using the formulae from the course lecture notes.

```python
def fourier_transform(x):
    """ 
    Performs a Fourier transform
    
    Parameters
    ----------
    x: np.ndarray 
        Array of real or complex values containing the signal
    
    Returns
    -------
    np.ndarray 
        Array of complex vaules containing the signal
    """
    N = x.shape[0]
    arr = np.asarray([np.sum(x * [np.exp(-2 * np.pi * 1j / N * k * n) for n in xrange(N)]) for k in xrange(N)])
    return arr
        
    
```

```python
def inverse_fourier_transform(X):
    """ 
    Performs an inverse Fourier transform
    
    Parameters
    ----------
    x: np.ndarray 
        Array of  complex values containing the signal
    
    Return
    ------
    np.ndarray 
        Array of complex vaules containing the signal
    """
    N = X.shape[0]
    arr = np.asarray([np.sum(X * [np.exp(2 * np.pi * 1j / N * k * n) for n in xrange(N)]) for k in xrange(N)]) * 1/N
    return arr
```
